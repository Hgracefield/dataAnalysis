#### 분석용 함수 사용하기

```python
import pandas as pd
import numpy as np
```

```python
data = [
    [1.4, np.nan],
    [7.1, -4.5],
    [np.nan, np.nan],
    [0.75, -1.3]
]
data
```

```python
df =\
    pd.DataFrame(
    data, 
    index = ['a', 'b', 'c', 'd'],
    columns = ['one', 'two'],
)
df
```

```python
df.info()
```

```python
# 열방향 합계 구하기
# pandas 에서는 nan 을 빼고 계산을 함
df.sum(
    axis='rows'
)
```

```python
# 행방향 합계 구하기
# pandas 에서는 nan 을 빼고 계산을 함
df.sum(
    axis='columns'
)
```

```python
# 특정 열의 합계 구하기
# df.one => series 1차원 데이터라서 방향을 줄 필요가 X
df.one.sum()
```

```python
# 특정 행의 합계 구하기
df.loc['b'].sum()
```

```python

df.mean(
    axis = 'rows',
    skipna=True
)
```

```python
# NaN 처리
df.one.fillna(value=df.one.mean(), inplace=True)
df
```

#### DataFrame Merging(병합) 
###### 키값 중심으로 데이터를 합치는것

```python
df1 = pd.DataFrame(
    {
        'key': list('bbacaab'),
        'data1' : range(7)
    }
)
df1
```

```python
df2 = pd.DataFrame({
    'key': list('abd'),
    'data2':range(3)
})
df2
```

```python
# 병합하기 : 기본값은 inner join
pd.merge(
    df1,
    df2,
    on='key'
) # 교집합. 
```

```python
# 병합하기 : outer join
pd.merge(
    df1,
    df2,
    on='key',
    how='outer'
) # 교집합. 
```

```python
# 병합하기 : left join
pd.merge(
    df1,
    df2,
    on='key',
    how='left'
) # 교집합. 
```

```python
# 병합하기 : right join
pd.merge(
    df1,
    df2,
    on='key',
    how='right'
) # 교집합. 
```

```python
df3 = pd.DataFrame(
    {
        'rkey': list('bbaaab'),
        'data1':range(6)
    }
)
df3
```

```python
df4 = pd.DataFrame(
    {
        'rkey': list('abd'),
        'data2':range(3)
    }
)
df4
```

```python
pd.merge(
    df3,
    df4,
    left_on='rkey',
    right_on='rkey',
    how='inner'
)
```

```python
cats = pd.cut(ages, bins)
cats
```

```python
cats.codes  
```

```python
cats.value_counts()
```

```python
# catogory 이름 정하기
group_names = ['Youth', 'YoungAdult', 'MiddleAge', 'Senior']
cat2 = pd.cut(ages, bins, labels = group_names)
cat2.value_counts()
```
