### 서울시 유가 정보 분석

```python
import pandas as pd

# file name 합치기
from glob import glob
```

```python
station_files = glob("../Data/oil/*xls")
station_files
```

```python
# 파일 이름 정렬하기
station_files = sorted(station_files)
station_files
```

```python
# !pip install xlrd  xls파일 불러올때 있어야함
```

```python
tmp_raw = []
for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
station_raw = pd.concat(tmp_raw)
```

```python
len(station_raw)
```

```python
station_raw.head()
```

```python
station_raw.info()
```

```python
# 필요항목 발췌
stations = pd.DataFrame(
    {
        'Oil_store':station_raw['상호'],
        '주소':station_raw['주소'],
        '가격':station_raw['휘발유'],
        '셀프' : station_raw['셀프여부'],
        '상표':station_raw['상표']
    }
)
stations.head()
```

```python
# 주소 컬럼에 서울특별시 또는 기타로 되어 있는 입력 부분 확인
set([resion.split()[0] for resion in stations.주소])  # collection에서 중복 제거 set()
```

```python
# 주소의 서울특별시와 서울시를 서울로 변경하기
stations.주소 = stations.주소.str.replace('서울특별시','서울')      
stations.주소 = stations.주소.str.replace('서울시','서울')   
stations.head()   
```

```python
stations.상표.unique()
```

```python
stations.reset_index(drop=True, inplace=True)
```

```python
stations.가격.unique()
```

```python
stations[stations.가격 == '-'] # 파일들을 합친자료로 인덱스 번호가 맞지 않아서 32를 사용 할 수 없음. 따라서 reset_index를 하고 선택하도록
```

```python
# 300번지 index 삭제

stations.drop(
    300,
    axis='rows',
    inplace=True
)
```

```python
stations.가격 == '-'
```

```python
stations.index == 8
```

```python
stations.info()
```

```python
stations.가격 = stations.가격.astype(int)
stations
```

```python
stations.info()
```

```python
stations['구']= [resion.split()[1] for resion in stations.주소]
stations.head()
```

```python
stations = \
    stations.sort_values(
        by='구',
    ).reset_index(drop=True)
stations.head()
```

```python
len(stations['구'].unique())
```

```python
# 강남구의 평균값



stations.pivot_table(
    values='가격',
    index='구',
    aggfunc='mean'
).head()
```

### 지도표시

```python
import json
import folium
```

```python
geo_path = "../Data/skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding = 'utf-8'))

map = folium.Map(
    location = [37.5502, 126.982],
    zoom_start=10.5
)

folium.Choropleth(
    geo_data=geo_str,
    data=stations_mean,
    columns=[stations_mean.index, '가격'],
    fill_color='PuRd',
    key_on= 'feature.id'
).add_to(map)

map
```
