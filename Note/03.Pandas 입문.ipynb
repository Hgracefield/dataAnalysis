```python
import pandas as pd # 행과 열이 있어서 기본적으로 2차원
```

```python
df = pd.DataFrame(
    [
        [  15, '남', '덕영중'   ],
        [  17, '여', '수리중'   ],
    ],
    index = ['준서', '예은'],
    columns=['나이', '성별', '학교']
)
df
```

```python
df.index = ['학생1', '학생2']
df.columns = ['연령', '남녀', '소속']
df
```

```python
# 열 이름중 연령 -> 나이
df.rename(
    columns={
        '연령':'나이'
    },
    inplace=True # 즉시 적용 ( 메모리에 )    
)
df
```

```python
# 열 이름중 남녀 -> 성별, 소속 -> 학교로 변경하기
df.rename(
    columns = {
        '남녀':'성별',
        '소속':'학교'
    },
    inplace=True
)
df
```

```python
# 행 이름을 변경하기 학생1 -> 준서, 학생2 -> 예은
df.rename(
    index={
        '학생1':'준서',
        '학생2':'예은'
    },
    inplace=True
)
df
```

#### 행 선택
- index 이름을 기준으로 선택할 경우에는 loc
- 정수형 위치 index를 선택할 경우에는 iloc

```python
# index가 서준인 행을 선택
print(df.loc['서준',:])
print(df.iloc[0,:]) # (index, )
```

```python
# loc와 iloc로 행이 서준, 우현인 데이터 추출하기

print(df.loc[['서준', '우현'], :])
print(df.iloc[0:1+1,:])
```
