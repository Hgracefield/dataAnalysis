- model year : 출시년도

```python
# 평균값
# print(df.mean()) # type 에러
print(df.mean(numeric_only=True))
print("-"*20)
print(df.mean(numeric_only=True)['mpg'])
print("-"*20)
print(df.mpg.mean())
print("-"*20)
print(df['mpg'].mean())
# mpg와 weight의 평균 보기
print("-"*20)
print(df[['mpg','weight']].mean())
```

```python
# 중앙값
print(df.median(numeric_only=True))
print("-"*20)
print(df.median(numeric_only=True)['mpg'])
print("-"*20)
print(df.mpg.median())
print("-"*20)
print(df['mpg'].median())
# mpg와 weight의 평균 보기
print("-"*20)
print(df[['mpg','weight']].median())
```
