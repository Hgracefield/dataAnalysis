#### Pandas 내장 그래프 도구

```python
# 남북한발전전력량.xlsx
df = pd.read_excel("../Data/남북한발전전력량.xlsx")
df
```

```python
# 남한, 북한 발전량 합계 데이터만 추출
df_sn = df.iloc[[0,5],2:] 
# [[0,5],1990:]

df_sn.index =[
        'South',
        'North'
    ]
df_sn
```

```python
# 데이터 타입 확인
df_sn.info()
```

```python
# obj -> int 로 변환
df_sn= df_sn.astype('int')
df_sn.info()
```

```python
import matplotlib.pyplot as plt
```

```python
# 선그래프 그리기
df_sn.plot()
plt.show()
```

```python
# 행과 열을 전치

tdf_sn = df_sn.transpose() # == df_sn.T
tdf_sn
```

```python
tdf_sn.plot()
plt.show()
```

```python
tdf_sn.plot(
    kind='bar' # 데이터 크기 비교할 때 사용
)
tdf_sn.plot
```

```python
# 히스토그램 : 빈도수 (숫자로 보고 싶으면 value_count)
tdf_sn.plot(
    kind='hist' 
)
tdf_sn.plot
```

#### 산점도 (산포도) : 상관계수


```python
df = pd.read_csv("../Data/auto-mpg.csv")
df.head()
```

```python
df[['weight', 'mpg']].corr()
```

```python
df.plot(
    x = 'weight',
    y = 'mpg',
    kind='scatter'
)
plt.show()
```

> 결측치(missing value) : 데이터의 분포도가 넓음                  
> 이상치(outliers) :  데이터가 튐 (max, min)

```python
# 상자수염그래프(박스플롯) => 이상치 찾기
df.mpg.plot(
    kind='box'
)
plt.show()
```

```python
# 상자수염그래프(박스플롯)
df[['mpg', 'weight']].plot(
    kind='box'
)
plt.show()
```

###### 겹치는 영역이 없어서 비교 할 수 없음(값 차이가 큼)
> 표준화 : 동기화 시키겠음                            
> 정규화 : 0~1 로 만들기
