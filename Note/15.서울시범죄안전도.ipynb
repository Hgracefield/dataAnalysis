```python
# 절도, 폭력의 dtype을 int로 변환
crime_anal_police['절도 발생'] = crime_anal_police['절도 발생'].str.replace(",",'').astype(int)
crime_anal_police['절도 검거'] = crime_anal_police['절도 검거'].str.replace(",",'').astype(int)
crime_anal_police['폭력 발생'] = crime_anal_police['폭력 발생'].str.replace(",",'').astype(int)
crime_anal_police['폭력 검거'] = crime_anal_police['폭력 검거'].str.replace(",",'').astype(int)
crime_anal_police.info()
```

```python
crime_anal_police.head()
```

```python
crime_anal = \
    crime_anal_police.pivot_table(
        index = '구별',
        aggfunc = 'sum'
    )
crime_anal.head()
```

```python
# 각 점죄별 검거율 구하기
crime_anal['강간검거율'] = crime_anal['강간 검거'] / crime_anal['강간 발생'] * 100
crime_anal['강도검거율'] = crime_anal['강도 검거'] / crime_anal['강도 발생'] * 100
crime_anal['살인검거율'] = crime_anal['살인 검거'] / crime_anal['살인 발생'] * 100
crime_anal['절도검거율'] = crime_anal['절도 검거'] / crime_anal['절도 발생'] * 100
crime_anal['폭력검거율'] = crime_anal['폭력 검거'] / crime_anal['폭력 발생'] * 100
crime_anal.head() # 검거는 지나온것도 있어서 100%가 나올 수 있음
```

```python
# 검거 건수는 삭제
del crime_anal['강간 검거']
del crime_anal['강도 검거']
del crime_anal['살인 검거']
del crime_anal['절도 검거']
del crime_anal['폭력 검거']
crime_anal
```

```python
# 율이 100% 넘는걸 100%로 수정하기
crime_anal[crime_anal.iloc[:,6:11] > 100] = 100
# 상동
# crimeList = ['강간검거율','강도검거율',	'살인검거율','절도검거율','폭력검거율']
# for i in crimeList:
#     crime_anal.loc[crime_anal[i] > 100, [i]] = 100

crime_anal
```

#### 데이터의 정규화(Normalization)
###### 데이터의 범위를 맞춤 = 정규화.
- 강도와 살인은 2자릿수 이나 절도와 폭력은 4자리수 입니다.
- 각 범죄 발생건수를 비슷한 범위에 놓고 비교한다.
- 각 항목의 최대값을 1로 계산하여 그 비중자체로 분석하는 것이 좋다.

```python
# 머신러닝 모듈
from sklearn import preprocessing
```

```python
# 정규화 
col = ['강간','강도','살인', '절도','폭력']
```

```python
# 데이터 설정
crime_anal[col]
```

```python
# 데이터 설정
x = crime_anal[col].values
x
```

```python
# !pip install scikit-learn
```

```python
# 최대값 최소값
min_max_scaler = preprocessing.MinMaxScaler()
min_max_scaler
```

###### fit -> 학습, transform -> 바꾸다.

```python
x_scaled = min_max_scaler.fit_transform(x.astype(float))
x_scaled  # 1이 max
```

```python
# 정규화된 범죄건수가 있는 새로운 데이터프레임 생성
crime_anal_norm = pd.DataFrame(
    x_scaled,
    columns=col,
    index = crime_anal.index
)
crime_anal_norm
```

```python
# 컬럼 추가
col2 = ['강간검거율','강도검거율','살인검거율', '절도검거율','폭력검거율']
crime_anal_norm[col2] = crime_anal[col2]
crime_anal_norm.head()
```

```python
# cctv 데이터와 합치기
result_CCTV = pd.read_csv("../Data/CCTV_result.csv", index_col='구별')
result_CCTV.head()
```

```python
crime_anal_norm[['인구수', 'CCTV']] = result_CCTV[['인구수', '소계']]
```

```python
import numpy as np
```

```python
crime_anal_norm['범죄'] = np.sum(crime_anal_norm[col], axis = 'columns')
crime_anal_norm.head()
```

```python
crime_anal_norm['검거'] = np.sum(crime_anal_norm[col2], axis = 'columns')
crime_anal_norm.head()
```
