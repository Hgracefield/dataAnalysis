### Series
- 데이터가 순차적으로 나열된 1차원 배열의 형태

```python
import pandas as pd
```

```python
# Dictionary -> Series
dict_data = {
    'a' : 1,
    'b' : 2,
    'c' : 3,
}
dict_data
```

```python
sr = \
    pd.Series(
        dict_data
    )
sr
```

---


```python
# List -> series

list_data = 
    {
        'r' : '2026-01-01',
        'm' : '3.14',
        'm' : 'abc',
        'm' : '100',
        'm' : 'True'
    }

```

```python
pd.Series(
    list_data
)
```

```python
# Tuple -> Series변환 후 index 변경하기

tuple_data = ('유비', '2020-10-20', '남', False)
sr = pd.Series(
    tuple_data,
    index=['이름', '생년월일', '성별', '학생여부']
)
sr
```

```python
tuple_data = ('유비', '2020-10-20', '남', False)
sr = pd.Series(
    tuple_data
)
sr.index = ['이름', '생년월일', '성별', '학생여부']
sr
```

```python
# iloc로 추출하기
sr.iloc[:2]

# loc로 추출하기
sr.loc[1,:]
```

#### Series의 연산

```python
student = pd.Series(
    {
        '국어':100,
        '영어':80,
        '수학':90
    }
)
student
```

```python
# 과목의 점수를 200으로 나누기

student / 200
```

```python
student1 = pd.Series(
    {
        '국어':100,
        '영어':80,
        '수학':90
    }
)
student1
```

```python
student2 = pd.Series(
    {
        '수학':80,
        '국어':90,
        '영어':80,
    }
)
student2
```

```python
# 덧셈
addition = student1 + student2
addition
```

```python
# 뺄셈
subtraction = student1 - student2
subtraction
```

```python
# 곱셈
multiplication = student1 * student2
multiplication
```

```python
# 나눗셈
division = student1 / student2
division
```

```python
# 데이터프레임 만들기
result = \
    pd.DataFrame(
        [addition, subtraction, multiplication, division],
        index=['덧셈', '뺄셈', '곱셈', '나눗셈']    
    )
result
```

```python
result = \
    pd.DataFrame(
        [addition, subtraction, multiplication, division],
        
    )
result.index = ['덯', '마','곱', '난']
result
```
