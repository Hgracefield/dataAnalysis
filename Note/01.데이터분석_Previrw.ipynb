#### 년도별 성별 births의 집계

###### 연도별 여자, 남자의 데이터를 만들어서 차트로 그리기


```python
# 컬럼별 값, 타입을 알아보자
names.info()
```

```python
# 1880
names.year.unique()
```

```python
### 남성과 여성의 연도별 births의 집계를 List로 구성하기
male = []
female = []


for year in sorted(names.year.unique()):
    female.append(int(names[(names['gender'] == 'F') & (names.year == year)].births.sum())) 
    male.append(int(names[(names['gender'] == 'M') & (names.year == year)].births.sum()))
print(len(female), len(male))
```

```python
len(female) == len(male)
```

```python
print(2014-1880+1)
```

```python
# DataFrame 만들기
total_births = \
pd.DataFrame(
    {
        'year' : sorted(names.year.unique()),
        'F' : female,
        'M' : male
    }
)
total_births.set_index('year', inplace=True) # inplace???
total_births.head()

```

---
#### 집계함수를 사용하지 않고 순수코딩으로 확인해 보기

```python
# 연도와 성별로 정렬
# 리스트는 sorted로 dataframe은 sort_~~
names2 = names.sort_values(by=['gender','year'] )
names2.head()
```

```python
female = []
male = []

for gender, births, year in zip(names2.gender, names2.births, names2.year):
    if gender == "F":
        try:
            female[year-1880]+=births
        except:
            female.append(births)
    else:
        try:
            male[year-1880]+=births
        except:
            male.append(births)
print(female)
print(male)

len(female) == len(male)
```

```python
# DataFrame 만들기
total_births2= \
pd.DataFrame(
    {
        'year' : sorted(names2.year.unique()),
        'F' : female,
        'M': male
    }
)
total_births2.set_index('year',inplace=True)
total_births2.head()

```
