#### 지도에 마크 표시하기
: 서울 시내 주요 대학교의 위치를 표시하기

```python
df = pd.read_excel("../Data/서울지역 대학교 위치.xlsx")
df.columns = ['학교명', '위도', '경도']
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# 서울 지도 만들기

seoul_map = folium.Map(
        location=[37.55, 126.98],  # 위도 , 경도
        zoom_start = 12,

)

for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    popup = folium.Popup(name, max_width=200, sticky=True)
    folium.Marker(
        [lat, lng],
        popup=popup
    ).add_to(seoul_map)


seoul_map
```

#### 지도에 원형 마커 표시

```python
# 서울 지도 만들기

seoul_map = folium.Map(
        location=[37.55, 126.98],  # 위도 , 경도
        zoom_start = 12,

)

for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    # popup = folium.Popup(name, max_width=200, sticky=True)
    folium.CircleMarker(
        [lat, lng],
        radius=10,
        color = 'brown',
        fill = True,
        fill_color = 'coral',
        fill_opacity=0.7
    ).add_to(seoul_map)


seoul_map
```

#### 지도 영역에 단계구분표(Choropleth Map)  표시하기

```python
import json
```

```python
df = pd.read_excel("../Data/경기도인구데이터.xlsx").set_index('구분')
df.head()
```

```python
df.columns = df.columns.map(str)
df.columns
```

```python
# 경기도 시군구 경계 정보
geo_path = "../Data/경기도행정구역경계.json"
try:
    geo_data = json.load(
        open(geo_path, encoding='utf-8')
    )
except:
    geo_data = json.load(
    open(geo_path, encoding='utf-8-sig')
    )
geo_data    
```

```python
# 결기도 지도
g_map = folium.Map(
    location = [37.5502,126.982],
    zoom_start=9
)

# 출력할 연도
year = '2007'

# Choropleth로 단계 구분 표시하기
folium.Choropleth(
    geo_data=geo_data, # 지도 경계
    data=df[year],
    columns=[df.index, df[year]],
    fill_color = "YlOrRd",
    fill_opacity=0.7,
    line_opacity=0.3,
    threshold_scale = [10000, 100000, 300000, 500000, 700000],
    key_on='feature.properties.name'
).add_to(g_map)

g_map
```
