```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# 행번호 0을 삭제하기
df.drop(
    index=0,
    inplace=True
)

# df.drop(
#     0,
#     axis = 'rows',
#     inplace=True
# )

```

```python
df.reset_index(
    drop=True,
    inplace=True
)
df.head()
```

```python
# NaN 처리 : NaN을 앞의 데이터로 채움

df.ffill(
   
    inplace=True
) # 앞에 있는 데이터를 채움. 바로 뒤의 데이터를 채우는 bfill
df.head()
```

---
#### 년도 컬럼의 데이터가 object로 되어있어 숫자로 변환

```python
df['1970'].unique()
```

```python
# 1970컬럼의 전체갯수
df['1970'].count()
```

```python
# -으로 등록되어 있는 데이터의 수
len(df[df['1970'] == '-'])
```

```python
# -이 아닌 등록되어 있는 데이터의 수
len(df[df['1970'] != '-'])
```

```python
# -을 0으로 대체.
df.replace(
    '-',
    0,
    inplace=True
)
df.head()
```

```python
df.info()

year_cols = df.columns[2:]
df[year_cols] = df[year_cols].apply(pd.to_numeric)
```

---
#### 서울에서 다른 지역으로 이동한 데이터만 추출

```python
df_seoul = \
    df[(df.전출지별 == '서울특별시') & (df.전입지별 != '서울특별시')]
df_seoul.head()
```

```python
# 전출지별 컬럼 삭제
df_seoul.drop(
    '전출지별',
    axis = 'columns',
    inplace=True
)
df_seoul.head()
```

```python
# 전입지별 -> 전입지로 변경
df_seoul.rename(
    {
        '전입지별' : '전입지'
    },
    axis='columns',
    inplace=True
)
df_seoul.head()
```

```python
# 전입지를 index로 변경

df_seoul.set_index(
    '전입지',
    inplace=True
)
df_seoul.head()
```

#### 그래프에 style 적용하기

```python
# Style 서식 지정
plt.style.use("ggplot")

# 그림 사이즈
plt.figure(
    figsize=(14, 5)
)
plt.plot(
    sr_one,
    marker = 'o',
    markersize=9
)
# 축 이름 
plt.xlabel('기간')
plt.ylabel('이동 인구수')
# 차트 타이틀
plt.title('부산 -> 서울 인구 이동')
# x 축 이름 이동 
plt.xticks(
    rotation=45
)
plt.yticks(
    rotation=15
)

# 범례
plt.legend(
    labels = ['부산 -> 서울'],
    loc='best'
)

# y축 범위 지정
plt.ylim(0, 800000)


# 주석 표시 - 화살표
plt.annotate(
    '',
    xy = (1990-1970, 620000), # 화살표의 머리 부분
    xytext=(1972-1970, 290000), # 화살표의 꼬리 부분
    xycoords='data',
    arrowprops=dict(arrowstyle='->', color='skyblue', lw=5) # 화살표 서식
)

# 조식표시 - 텍스트
plt.annotate(
    '인구이동 증가 (1970~1995)', # 텍스트 입력
    xy=(1980-1970, 400000), # 텍스트 위치 기준점
    rotation=23, # 텍스트 회전 각도
    va='baseline', # 텍스트 상하 정렬
    ha='center', # 텍스트 좌우 정렬
    fontsize=12
)


# 주석 표시 - 화살표 (떨어지기)
plt.annotate(
    '',
    xy=(2017-1970, 450000),
    xytext=(2000-1970, 580000),
    xycoords='data',
    arrowprops=dict(arrowstyle='->', color='olive', lw=5)
)

plt.annotate(
    '인구이동 감소(1995-2017)',
    xy=(2009-1970, 530000),
    rotation=-9,
    va='baseline',
    ha='center',
    fontsize=12
)

# 1995년 데이터 표시
plt.annotate(
    f"{sr_one['1995']:,}명",
    xy=(1995-1970, sr_one['1995'] + 31000),
    rotation = 0,
    va = 'baseline',
    ha = 'center',
    fontsize = 12
)

plt.annotate(
    f"{sr_one['2017']:,}명",
    xy=(2017-1970, sr_one['2017']-50000),
    rotation=0,
    va='baseline',
    ha='center',
    fontsize=12
)

# 1995년 마커 모양 추가하기 
plt.scatter(
    1995-1970,
    sr_one['1995'],
    color = 'red',
    s=350,
    marker='+',
)

plt.show()
```
